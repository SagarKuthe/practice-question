```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import fetch_california_housing
boston_data=fetch_california_housing()
```

```python
print(boston_data.keys())
```

```python
boston_data.DESCR
```

```python
boston=pd.DataFrame(boston_data.data,columns=boston_data.feature_names)
boston.head()
```

```python
boston.isnull().sum()
```

```python
boston.shape
```

```python
boston.info()
```

```python
boston.drop(columns=['Latitude','Longitude'])
```

```python
boston['MHV']=boston_data.target
```

```python
boston
```

```python
#plotting the histogram
sns.distplot(boston['MHV'],bins=20)
plt.show()
```

```python
corln=boston.corr().round(2)
```

```python
sns.heatmap(data=corln,annot=True)
```

```python
plt.figure(figsize=(20,5))
feature=['AveBedrms','Population']
targ=boston['MHV']
for i,col in enumerate(feature):
    plt.subplot(1,len(feature),i+1)
    x=boston[col]
    y=targ
    plt.scatter(x,y,marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MHV')
    
    
```
