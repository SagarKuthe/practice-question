```python
#calculating the mean of production Column
prod_mean=df1['Production'].mean()
prod_mean
```

```python
#calculating the standard Deviation of Production Column
prod_std=df1['Production'].std()
prod_std
```

```python
#Calculating the sum of null values of production Column
null_sum=df1['Production'].isnull().sum()
null_sum
```

```python
# Replacing the NaN values with Generated random values
rand_values=np.random.randint(prod_mean-prod_std,prod_mean+prod_std,size=null_sum)
prod_slice=df1['Production'].copy()
prod_slice[np.isnan(prod_slice)]=rand_values
df1['Production']=prod_slice
df1['Production']=df1['Production'].astype(int)
```

```python
df4=pd.read_csv(r"C:\Users\admin\Downloads\state_wise_crop_production.csv")
df4
```

```python
df4.columns=["Crop","State","Cost of Cultivation (`/Hectare) A2+FL","Cost of Cultivation (`/Hectare) C2","Cost of Production (`/Quintal) C2","Yield"]
```

```python
df4.columns
```

```python
df4.info()
```

```python
df4.describe()
```

```python
df4.corr()
```

```python
sns.boxplot(x=df4["Cost of Production (`/Quintal) C2"])
```

```python
x=df4[["Cost of Cultivation (`/Hectare) A2+FL","Cost of Cultivation (`/Hectare) C2","Cost of Production (`/Quintal) C2"]]
y=df4["Yield"]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
np.sqrt(mse)
```
